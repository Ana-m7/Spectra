```python
import pandas as pd
```

```python
df = pd.read_csv('Toddler Autism dataset July 2018.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
# check for missing values
print("Missing values:")
print(df.isnull().sum())

print("\nTarget column distribution:")
print(df['Class/ASD Traits '].value_counts())

print("\nAge range in months:")
print(df['Age_Mons'].describe())
```

```python
print("Sex distribution:")
print(df['Sex'].value_counts())

print("\nFamily history of ASD:")
print(df['Family_mem_with_ASD'].value_counts())

print("\nJaundice:")
print(df['Jaundice'].value_counts())

print("\nA1-A10 question averages (how often answered 1):")
q_cols = ['A1','A2','A3','A4','A5','A6','A7','A8','A9','A10']
print(df[q_cols].mean().round(2))
```
